```python
import pandas as pd
import re
```

```python
gen_data = pd.read_csv("Plant_1_Generation_Data.csv")
gen_data = gen_data.groupby(gen_data['DATE_TIME']).sum().drop('TOTAL_YIELD', 1)
gen_data.index = gen_data.index.map(lambda x: re.sub('(\d{2})-(\d{2})-(\d{4}) (\d{2}:\d{2})', '\\3-\\2-\\1 \\4', x))
gen_data.sort_index(inplace=True)
gen_data
```

```python
weather_data = pd.read_csv("Plant_1_Weather_Sensor_Data.csv")
weather_data.set_index('DATE_TIME', inplace=True)
weather_data.index = weather_data.index.map(lambda x: x[:-3])
weather_data.sort_index(inplace=True)
weather_data.drop('PLANT_ID', 1, inplace=True)
weather_data
```

```python
data = gen_data.join(weather_data)
data['AMBIENT_TEMPERATURE'] = data['AMBIENT_TEMPERATURE'].interpolate()
data['MODULE_TEMPERATURE'] = data['MODULE_TEMPERATURE'].interpolate()
data['IRRADIATION'] = data['IRRADIATION'].interpolate()
data['DC_POWER'] = data['DC_POWER'].map(lambda x: 1 if x > 0 else 0)
data
```

### Feature engineering using standard deviation

```python
factor = 2
upper_lim = data['IRRADIATION'].mean() + data['IRRADIATION'].std() * factor
lower_lim = data['IRRADIATION'].mean() - data['IRRADIATION'].std() * factor

print(upper_lim)
print(lower_lim)

data = data[(data['IRRADIATION'] < upper_lim) & (data['IRRADIATION'] > lower_lim)]
data
```

### Feature engineering using percentiles

```python
upper_lim = data['AMBIENT_TEMPERATURE'].quantile(0.95)
lower_lim = data['AMBIENT_TEMPERATURE'].quantile(0.05)

print(upper_lim)
print(lower_lim)

data = data[(data['AMBIENT_TEMPERATURE'] < upper_lim) & (data['AMBIENT_TEMPERATURE'] > lower_lim)]
data
```

```python
upper_lim = data['MODULE_TEMPERATURE'].quantile(0.95)
lower_lim = data['MODULE_TEMPERATURE'].quantile(0.05)

print(upper_lim)
print(lower_lim)

data = data[(data['MODULE_TEMPERATURE'] < upper_lim) & (data['MODULE_TEMPERATURE'] > lower_lim)]
data
```

### Feature selection using wrapper method (recursive feature elimination)

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier

x_train = data.drop(['DC_POWER', 'AC_POWER', 'SOURCE_KEY', 'PLANT_ID'], 1)
estimator = RandomForestClassifier(random_state = 42)
selector = RFE(estimator, n_features_to_select=2, step=1)
selector = selector.fit(x_train, data['DC_POWER'])

rfe_mask = selector.get_support()
new_features = [] 

for bool, feature in zip(rfe_mask, x_train.columns):
    if bool:
        new_features.append(feature)
new_features
```
