```python
import pandas as pd
```

# Missing values
Missing values project based on weather data of Vancouver, Canada between 1st October 2012 and 30th November 2017

```python
# Hourly humidity data from https://www.kaggle.com/selfishgene/historical-hourly-weather-data
humidity = pd.read_csv('humidity.csv')
humidity
```

```python
temperature = pd.read_csv('temperature.csv')
pressure = pd.read_csv('pressure.csv')
wind_speed = pd.read_csv('wind_speed.csv')
```

```python
humidity_vancouver = humidity[['datetime', 'Vancouver']].rename(columns = {'Vancouver':'humidity'})
temperature_vancouver = temperature[['datetime', 'Vancouver']].rename(columns = {'Vancouver':'temperature'})
pressure_vancouver = pressure[['datetime', 'Vancouver']].rename(columns = {'Vancouver':'pressure'})
wind_speed_vancouver = wind_speed[['datetime', 'Vancouver']].rename(columns = {'Vancouver':'wind_speed'})
ht_vancouver = pd.merge(humidity_vancouver, temperature_vancouver, on='datetime')
pw_vancouver = pd.merge(pressure_vancouver, wind_speed_vancouver, on='datetime')
weather = pd.merge(ht_vancouver, pw_vancouver, on='datetime')
weather
```

### List-based deletion

```python
list_deleted_weather = weather.dropna()
list_deleted_weather
```

### Pair-based deletion

```python
pair_deleted_weather = weather.dropna(subset=['humidity', 'temperature'])
pair_deleted_weather
```

### Deletion of column

```python
print("Humidity:", weather['humidity'].isna().sum())
print("Temperature:", weather['temperature'].isna().sum())
print("Pressure:", weather['pressure'].isna().sum())
print("Wind Speed:", weather['wind_speed'].isna().sum())
```

```python
dc_weather = weather.copy()
dc_weather.drop('pressure', axis=1, inplace=True)
dc_weather
```

### LOCF

```python
locf_weather = weather.copy()
locf_weather.fillna(method='ffill')
```

### NOCB

```python
nocb_weather = weather.copy()
nocb_weather.fillna(method='bfill')
```

### Linear Interpolation

```python
li_weather = weather.copy()
li_weather['humidity'] = li_weather['humidity'].interpolate()
li_weather
```
