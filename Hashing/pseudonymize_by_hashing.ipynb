```python
import pandas as pd
import hashlib
```

```python
# Hash the input_text using SHA3 and return 256-bit hash-value
def hash_string(input_text):
    encoded_input = input_text.encode()
    sha_obj = hashlib.sha3_256(encoded_input)
    return sha_obj.hexdigest()
    
hash_string("abc")
```

```python
# Import data
data = pd.read_csv('orders.csv')
data
```

```python
import secrets
salt = secrets.token_urlsafe(16)
salt
```

```python
# Hash column name
data['name'] = data['name'].apply(lambda x: hash_string(x + salt))
data
```

```python
# Export back to csv
data.to_csv("pseudonymized_order.csv")
```
